## Ordinal Encoding

```python
X_train['Package Type'].unique()
```

```python
pt=['Budget','Standard', 'Premium','Deluxe','Luxury']
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordi=OrdinalEncoder(categories=[pt])
ordi.fit(X_train[['Package Type']])
```

```python
pd.DataFrame(ordi.transform(X_train[['Package Type']]))
```
